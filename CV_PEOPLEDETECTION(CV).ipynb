```python
import cv2
import numpy as np
from ultralytics import YOLO
from google.colab.patches import cv2_imshow


model = YOLO("/content/best.pt")


source = "/content/download.jpg"

results = model.predict(source, device="cpu", conf=0.25)

img = cv2.imread(source)
h, w, _ = img.shape
frame_area = h * w

person_count = 0

for box in results[0].boxes:
    cls = int(box.cls[0])
    if model.names[cls] == "person":
        person_count += 1
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        area = (x2 - x1) * (y2 - y1)
        total_person_area += area
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)


density = total_person_area / frame_area

if person_count >= 6 and density >= 0.25:
    alert_text = "‚ö† CROWD DETECTED"
    color = (0, 0, 255)
else:
    alert_text = "Normal Crowd"
    color = (0, 255, 0)

cv2.putText(img, alert_text, (20, 40),
            cv2.FONT_HERSHEY_SIMPLEX, 1.1, color, 3)

cv2.putText(img, f"People: {person_count}", (20, 80),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 0), 2)

cv2.putText(img, f"Density: {density:.2f}", (20, 120),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 0), 2)


cv2_imshow(img)
```

```python
from ultralytics import YOLO
import cv2


model = YOLO("/content/best.pt")

results = model.predict(
    source="/content/polling.jpg",
    device="cpu",
    conf=0.15,
    imgsz=640,
    save=True,
    verbose=False
)

person_count = len(results[0].boxes)


estimated_count = person_count + 1 if person_count >= 4 else person_count

print(f" Detected voters in queue: {person_count}")
print(f" Estimated actual voters: {estimated_count}")


if estimated_count <= 5:
    status = "LOW"
    action = "No action required"
elif estimated_count <= 10:
    status = "MEDIUM"
    action = "Monitor queue"
else:
    status = "HIGH"
    action = "Deploy extra staff / voting machines"

print(f"‚úÖ Queue Status: {status}")
print(f"‚ö†Ô∏è Action: {action}")


results[0].show()
```
